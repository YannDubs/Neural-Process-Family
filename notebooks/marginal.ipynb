```python
def sarle(out, axis=0):
    k = scipy.stats.kurtosis(out, axis=axis, fisher=True)
    g = scipy.stats.skew(out, axis=axis)
    n = out.shape[1]
    denom = k + 3*(n-1)**2/((n-2)*(n-2))
    return (g**2 +1 )/denom
```
